## Research Question

Has the production of Slice of Life anime passed Shounen anime from 1990 to 2025 and if so when did this trend start? The trend is defined as the annual ratio of genre-specific releases worldwide to total releases.

## Background and Prior Work

Popular media is a reflection of the interests of the majority and changes over time and place through the influence of culture and environment. Trends can arise when certain genres attract a significant following and creators appeal to this interest, leading to increased production in media with common themes. In terms of anime, two genres that have had a consistently large following and interest are shonen and slice-of-life. Many iconic anime fall under the shonen genre such as Naruto and Dragon Ball and are usually characterized by action and fantasy themes. Similarly, figures from slice-of-life anime such as Lucky Star and Love Live have become easily recognizable from their long-standing popularity and are characterized by lighthearted and relatable stories. With shonen and slice-of-life being popular genres in Japanese media for a long time, we wanted to better understand any potential changes in production between anime of these two genres over the years.

Many shonen series first find their start by having a chapter published in the Weekly Shonen Jump magazine before being compiled into a full manga. The target audience of this magazine is usually teen and young adult boys, which is seen through the frequent use of male protagonists who reflect this audience. The article “The Rise and Fall of Weekly Shonen Jump: A Look at the Circulation of Weekly Jump” focuses on the amount of sales of the Weekly Shonen Jump magazine from 1970 to 2005 and takes note of its peak popularity and when it begins to decline <a href="#ref1">[1]</a>. Observing sales of the magazine, it began its height of popularity in 1993 but “…between 1995 and 2005, while other magazines' sales stayed fairly constant, Shonen Jump's circulation decreased dramatically” <a href="#ref1">[1]</a>. At the same time, competing magazines like Weekly Shonen Magazine also found a slight decline in sales around a similar time period <a href="#ref1">[1]</a>. Comparing this change in sales to the titles in the editions of this time, the decrease in sales coincides with the ending of popular series like Dragon Ball <a href="#ref1">[1]</a>. This study was done in 2007 and only observes sales up to 2005. Additionally, the data solely focuses on the popularity of titles being published through this particular magazine, which may not reflect titles in other magazines or other platforms like television.

In “Trends of Fiction in 2000s Japanese Pop Culture” published in the Electronic Journal of Contemporary Japanese Studies, Motoko Tanaka studies the fall of “the fictional age” (kyokō no jidai) and the beginning of “the age of impossibility” zeitgeists in Japan, along with how that change correlated with the rise and fall of certain anime genres popular among teen-young adult Japanese men <a href="#ref2">[2]</a>. Tanaka argues that the 1970–1995 “fictional age,” which was heralded in by the failure of the Anpo and Zenkyōtō student activism movements, was defined by the desire of young Japanese men to “…seek ideals not in political movements and changes in the real world, but rather in imagined settings such as theme parks and in fictional characters” <a href="#ref2">[2]</a>. While the article doesn’t directly mention shonen anime being popular, it can be logically inferred that anime with strong, morally just male protagonists in fantasy settings fit this cultural desire. Tanaka then discusses the fall of the “fictional age” and the beginning of "the age of impossibility,” influenced by catastrophes such as the 1995 Kobe Earthquake, the Aum sarin gas attack, and 9/11, as well as Japan’s ‘lost decade’ <a href="#ref2">[2]</a>. During this period, more optimistic genres such as traditional shonen saw a decline in prominence, while darker genres such as Sekaikei and Sabaibukei (death game narratives) gained attention <a href="#ref2">[2]</a>. Tanaka also notes that, during the late 2000s, Nichijōkei (slice-of-life) anime began to surge in popularity <a href="#ref2">[2]</a>. Although the paper focuses more on cultural analysis than statistical evidence and is limited mainly to the 1990s–2000s, it still identifies clear genre trends and their cultural context, which helped inform our hypothesis.

A more recent study that focused more on data analysis was performed by Research and Markets, with their results being presented in the article “United States Anime Genres Market Analysis Report 2025: Rise of Isekai and Fantasy Genres - Competition, Trends, Forecasts & Opportunities, 2020–2030” <a href="#ref3">[3]</a>. This article summarizes current trends in the anime industry and predicts future trends and growth, with a focus on the American market. It states that the U.S. anime market has grown significantly due to the rise of streaming services, merchandise growth, and expanding demographic appeal, with the market expected to double by 2030 <a href="#ref3">[3]</a>. It also states that, while shonen and slice-of-life anime continue to be extremely popular genres, they are no longer at the very top. Instead, fantasy and particularly isekai genres are now dominant, influenced by the success of series like Sword Art Online <a href="#ref3">[3]</a>. While the article provides market growth data and industry projections, it does not provide detailed genre-specific statistics and is limited to the 2020s–2030s timeframe.

In general, it seems that, compared to our planned study, most previous studies are either narrower in their temporal focus (we plan on analyzing trends from the 1960s to the 2020s) and/or do not include explicit long-term quantitative analyses and visualizations.

## Hypothesis


We believe that Slice of Life anime production experienced explosive growth while Shounen production experienced a plateau/drop starting around the year 2000. Shounen was extremely popular around 1990 to the early 2000s. However, starting in the year 2000 we believe that simpler storylines that follow easily digestible characters and lack action have become more popular. We expect to observe an inverse relationship between the production rate of Shounen and Slice of Life worldwide with an intersection point around the year 2000.

### Anime Production Dataset (1969~now)

For this anime dataset, we are focusing on the columns ‘anime_id’, ‘title’, ‘type’, ‘episodes’, ‘duration’, ‘genres’, ‘start_date’, and ‘end_date’. The ‘title’ column shows the official untranslated titles of each series in this dataset and romanized the Japanese titles if applicable. In the ‘duration’ column, the length of each episode is listed in seconds as an integer. Titles categorized as “TV” under ‘type’ are usually between 720-1440 seconds in length while movies tend to be 3600 seconds or longer. Under the ‘genres’ column, each title is categorized with a list of all the genres that are applicable. Titles can have at least one or more genres in its list. We will only be comparing titles where either the genres shonen or slice-of-life appear with no overlap. The ‘start_date’ column lists the initial date that the title began to air in year-month-day format. The ‘end_date’ is formatted similarly with a date in year-month-day format. A title will not have a value under this column if the series is ongoing and has not been finished airing.

The data in this dataset is updated daily from the website MyAnimeList which was first created in 2004. This website is mostly used by English-speaking users and is paired with streaming services popular in the U.S. like Crunchyroll and Hidive. As a result, the availability and data regarding older titles prior to the website’s creation may be less available. Additionally, user based data may tend to reflect Western audiences who use the website itself as well as the streaming services partnered with them while leaving out titles that are only available in Japan.

##### Process genres
The genres column is in the form "genre 1;genre 2;genre 3;..." Since we only care about the slice-of-life and shounen anime, we create two binary columns "is_slice_of_life" and "is_shounen" from the genres column. If "Slice of life" is contained the genre, we will set is_slice_of_life to True, otherwise, False. If "Shounen" is contained the genre, we will set is_shounen to True, otherwise, False. There are a few anime that are both slice-of-life and shounen. Since there are only a few of them, we remove them from our analysis.

```python
genres_to_analyze = (clean_data.SHOUNEN_TAG, clean_data.SLICE_OF_LIFE_TAG)
df = clean_data.gen_binary_categories(df=df, tags_to_check=genres_to_analyze)
slice_of_life_and_shounen_cnt = df[df['is_slice_of_life'] & df['is_shounen']].shape[0]
print(f"There are {slice_of_life_and_shounen_cnt} ({round(slice_of_life_and_shounen_cnt/df.shape[0], 2)}%) entries that are both slice of life and shounen")
```

```python
df = clean_data.filter_multicategory_anime(df=df, tags_to_check=genres_to_analyze)
df.head(5)
```

##### Process dates
We convert the start and end dates from string to datetime objects. Then, we added start_year, end_year, and aired_days as floats since we will mainly analysis the production trend across years. For shows that are still airing, we calculate its end year and aired day by using ref_data (2026-02-01 for this analysis) as its last aired date.

```python
df = clean_data.add_airing_date_features(df=df)
df.head(5)
```
